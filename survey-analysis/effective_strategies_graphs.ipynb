```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# Sample DataFrame structure
data = {
    'Category': ['Beauty', 'Beauty', 'Beauty', 'Beauty', 'Beauty'] * 2,
    'Attribute': ['natural/parks', 'rivers/canals', 'urban', 'beaches', 'industrial'] * 2,
    'Cluster': ['scenic'] * 5 + ['urban'] * 5,
    'Value': [1.5, 1.2, 0.5, 0.8, -0.5, -1.2, -0.8, -0.5, -1.0, -1.5],
    'Significance': ['***', '***', '***', 'ns', '***', '***', '***', '***', 'ns', '***']
}

df = pd.DataFrame(data)

# Create a custom function for significance annotations
def add_stat_annotations(ax, data, attribute_col, value_col, significance_col):
    grouped_data = data.groupby(attribute_col)
    for idx, (attr, group) in enumerate(grouped_data):
        cluster_values = group[value_col]
        sig_text = group[significance_col].iloc[0]
        y_pos = cluster_values.max() + 0.2
        ax.text(idx, y_pos, sig_text, ha='center', color='black', fontsize=10)

# Plotting
sns.set(style="whitegrid")
fig, axes = plt.subplots(1, 1, figsize=(10, 6))

# Boxplot
sns.boxplot(
    data=df, x='Attribute', y='Value', hue='Cluster',
    palette={'scenic': 'skyblue', 'urban': 'salmon'},
    ax=axes
)

# Add statistical significance annotations
add_stat_annotations(axes, df, 'Attribute', 'Value', 'Significance')

# Title and labels
axes.set_title('Beauty', fontsize=14)
axes.set_ylabel('Value', fontsize=12)
axes.set_xlabel('')

# Custom legend
legend_elements = [
    Patch(facecolor='skyblue', edgecolor='black', label='Cluster 1: Scenic'),
    Patch(facecolor='salmon', edgecolor='black', label='Cluster 2: Urban')
]
axes.legend(handles=legend_elements, title='Cluster', loc='upper right')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dummy data
np.random.seed(42)
categories = ['Category 1', 'Category 2', 'Category 3']
cluster1_data = [
    np.random.normal(1.0, 0.3, 10),
    np.random.normal(0.8, 0.2, 10),
    np.random.normal(0.5, 0.3, 10),
]
cluster2_data = [
    np.random.normal(-0.5, 0.4, 10),
    np.random.normal(-1.0, 0.5, 10),
    np.random.normal(0.0, 0.4, 10),
]

# Define positions with a gap between boxes
positions_cluster1 = [x - 0.2 for x in range(len(categories))]
positions_cluster2 = [x + 0.2 for x in range(len(categories))]

# Create the plot
plt.figure(figsize=(10, 6))
plt.gca().set_facecolor('#F5F5F5')  # Light grey background

# Plot Cluster 1
plt.boxplot(
    cluster1_data,
    positions=positions_cluster1,
    widths=0.3,  # Adjust box width
    patch_artist=True,
    boxprops=dict(facecolor='skyblue', color='black'),
    medianprops=dict(color='black'),
)

# Plot Cluster 2
plt.boxplot(
    cluster2_data,
    positions=positions_cluster2,
    widths=0.3,  # Adjust box width
    patch_artist=True,
    boxprops=dict(facecolor='salmon', color='black'),
    medianprops=dict(color='black'),
)

# Overlay diamonds (means) with higher zorder
means_cluster1 = [np.mean(data) for data in cluster1_data]
means_cluster2 = [np.mean(data) for data in cluster2_data]

# Add diamonds for Cluster 1
plt.scatter(
    positions_cluster1, means_cluster1,
    color='blue', marker='D', s=50, label='Mean Cluster 1', zorder=5
)

# Add diamonds for Cluster 2
plt.scatter(
    positions_cluster2, means_cluster2,
    color='red', marker='D', s=50, label='Mean Cluster 2', zorder=5
)

# Add vertical white lines between categories
for x in range(len(categories) - 1):
    plt.axvline(x=x + 0.5, color='white', linewidth=1)

# Add top-title for each category
for i, category in enumerate(categories):
    y_max = 2.1  # Adjusted height for the text placement
    plt.text(
        x=i, y=y_max, s='Top-title', ha='center', va='bottom',
        fontsize=10, color='black', weight='bold'
    )

# Customize axes
plt.xticks(range(len(categories)), categories, fontsize=10)  # Set category names
plt.yticks(np.arange(-2, 2.5, 0.5), fontsize=10)  # Y-axis ticks
plt.ylim(-2, 2)  # Set y-axis limits
plt.title('Comparison of Clusters Across Categories', fontsize=14, weight='bold')

# Clean up axes
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.grid(color='white', linewidth=1, axis='y', alpha=1)

# Add a legend for the diamonds
plt.legend(loc='upper left', fontsize=10)

# Show plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# Dummy data for each category group
category_groups = [
    ['Beauty', 'Air Pollution'],                # Group 1: 2 categories
    ['Running Ground', 'Noise Levels', 'Crowds'],  # Group 2: 3 categories
    ['Traffic', 'Safety', 'Obstacles'],         # Group 3: 3 categories
    ['Category 1', 'Category 2'],              # Group 4: 2 categories
    ['X', 'Y', 'Z', 'W'],                      # Group 5: 4 categories
    ['A', 'B', 'C'],                           # Group 6: 3 categories
]



# Generate dummy data for each category group
np.random.seed(42)
data_by_group = [
    [np.random.normal(0.5, 0.3, 10), np.random.normal(-0.3, 0.2, 10)],    # Group 1
    [np.random.normal(1.0, 0.3, 10), np.random.normal(0.8, 0.2, 10), np.random.normal(0.2, 0.3, 10)],  # Group 2
    [np.random.normal(0.5, 0.2, 10), np.random.normal(-0.8, 0.4, 10), np.random.normal(0.3, 0.2, 10)], # Group 3
    [np.random.normal(0.1, 0.2, 10), np.random.normal(-0.2, 0.3, 10)],    # Group 4
    [np.random.normal(0.8, 0.4, 10), np.random.normal(0.5, 0.2, 10), np.random.normal(-0.1, 0.3, 10), np.random.normal(-0.3, 0.2, 10)],  # Group 5
    [np.random.normal(1.0, 0.3, 10), np.random.normal(0.5, 0.2, 10), np.random.normal(-0.2, 0.3, 10)],  # Group 6
]

# Create a figure with GridSpec
fig = plt.figure(figsize=(12, 12))
gs = GridSpec(3, 2, figure=fig)  # 3 rows, 2 columns layout

# Colors for the clusters
colors = ['skyblue', 'salmon']

# Iterate through category groups and add a subplot for each
for i, (categories, data) in enumerate(zip(category_groups, data_by_group)):
    ax = fig.add_subplot(gs[i // 2, i % 2])  # Determine position in the grid
    cluster1_data = data[:len(categories)]
    cluster2_data = [np.random.normal(-0.5, 0.4, 10) for _ in range(len(categories))]

    # Define positions for boxplots within each category
    positions_cluster1 = [x - 0.2 for x in range(len(categories))]
    positions_cluster2 = [x + 0.2 for x in range(len(categories))]

    # Plot Cluster 1
    ax.boxplot(
        cluster1_data,
        positions=positions_cluster1,
        widths=0.3,
        patch_artist=True,
        boxprops=dict(facecolor=colors[0], color='black'),
        medianprops=dict(color='black'),
    )

    # Plot Cluster 2
    ax.boxplot(
        cluster2_data,
        positions=positions_cluster2,
        widths=0.3,
        patch_artist=True,
        boxprops=dict(facecolor=colors[1], color='black'),
        medianprops=dict(color='black'),
    )

    # Add diamonds for the means
    means_cluster1 = [np.mean(data) for data in cluster1_data]
    means_cluster2 = [np.mean(data) for data in cluster2_data]

    # For Cluster 1
    ax.scatter(
        positions_cluster1, means_cluster1, 
        facecolors='none', edgecolors='black', 
        marker='D', s=50, zorder=5
    )

    # For Cluster 2
    ax.scatter(
        positions_cluster2, means_cluster2, 
        facecolors='none', edgecolors='black', 
        marker='D', s=50, zorder=5
    )

    # Add vertical white lines to separate categories
    for x in range(len(categories) - 1):
        ax.axvline(x=x + 0.5, color='white', linewidth=1)

    # Add top-title for each category
    for j, category in enumerate(categories):
        y_max = 2.1  # Adjusted height for the text placement
        ax.text(
            x=j, y=y_max, s='Top-title', ha='center', va='bottom',
            fontsize=10, color='black', weight='bold'
        )

    # Customize axes
    ax.set_xticks(range(len(categories)))
    ax.set_xticklabels(categories, fontsize=9, rotation=20)
    ax.set_yticks(np.arange(-2, 2.5, 0.5))
    ax.set_ylim(-2, 2)
    ax.grid(color='white', linewidth=1, axis='y', alpha=1)
    ax.set_facecolor('#F5F5F5')

    # Clean up spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)

# Adjust layout
fig.suptitle('Boxplot Comparison Across Category Groups', fontsize=16, weight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.97])  # Leave space for title
plt.show()
```
